# Dimensionality Reduction
It is used for reducing the number of features or input variables while retaining as much data information as possible.

## Why Dimensionality Reduction is Used?

| Reason | Explanation |
| :--- | :--- |
| **Avoid overfitting** | Too many features can cause the model to perform well on training data but fail on new data. |
| **Improve Model Performance** | Fewer inputs often mean faster training and better generalization. |
| **Simplify Visualization** | Humans can understand 2D or 3D plots; reducing dimensions helps visualize patterns. |
| **Reduce Storage and Computation** | Less memory and faster computation when handling large datasets. |
| **Handle multicollinearity** | Removes redundant features that are highly collinear. |

### 1. PCA (Principal Component Analysis)

#### Definition
PCA is a linear dimensionality reduction technique that:

*   Transforms original features into a new set of uncorrelated features called **Principal Components**.
*   Captures the maximum variance present in the original data—the first few components retain most of the information.

#### Step-by-Step Detailed Explanation of PCA:
1. **Standardize the data**: Scale the features so they have a mean of 0 and a variance of 1.
2. **Calculate the covariance matrix**: Identify how variables vary from the mean with respect to each other.
3. **Compute eigenvalues and eigenvectors**: Determine the principal components of the data.
4. **Sort eigenvalues and eigenvectors**: Order them to identify the most significant components.
5. **Select top k Principal Components**: Choose the components that explain the most variance.
6. **Project the data**: Transform the original data into the new k-dimensional subspace.

#### Covariance Formulas

**Population Covariance formula:**
Used when you have the entire dataset.
$$Cov(x,y) = \frac{\sum (x_i - \mu_x)(y_i - \mu_y)}{N}$$

**Sample Covariance formula:**
Used when working with a subset/sample of the data.
$$Cov(x,y) = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{n-1}$$

```python
import pandas as pd
df = pd.read_csv('breast-cancer-wisconsin-data.csv')
print(df.info())
print(df.head(10))
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Drop non-predictive columns (id) and separate the target (diagnosis)
X = df.drop(['id', 'diagnosis'], axis = 1)

# 2. Initialize and apply the scaler
ss = StandardScaler()
X_scaled = ss.fit_transform(X)

# 3. Convert back to a DataFrame for better readability
# This helps keep track of column names after transformation
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

display(X_scaled.head(10))
```

```python
y = df['diagnosis']

y = y.map({'M': 1, 'B': 0})
y.head(10)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
#importing linear regression
from sklearn.linear_model import LogisticRegression
#importing metrics
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(X_scaled)

X_pca = pd.DataFrame(X_pca)
X_pca.head(10)
```

```python
variance_df = pd.DataFrame({'Principal Component': [f'PC{i + 1}' for i in range(len(pca.explained_variance_ratio_))],
                          'Variance Explained': pca.explained_variance_ratio_,
                           'Cumulative Variance': pca.explained_variance_ratio_.cumsum()
})
variance_df
```

```python
pca_95 = PCA(n_components=0.95)
X_pca_95 = pca_95.fit_transform(X_scaled)

X_pca_95 = pd.DataFrame(X_pca_95)

print("Original Features:", X.shape[1])
print("Selected PCs:", X_pca_95.shape[1])
```

```python
loadings = pd.DataFrame(
    pca.components_.T,
    columns = [f'PC{i+1}' for i in range(30)],
    index = X.columns
)

print(loadings[['PC1', 'PC2', 'PC3']])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_pca_95,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```

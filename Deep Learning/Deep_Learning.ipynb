## Artificial Neural Networks: Comprehensive Study Guide

### 1. Fundamental Terminology
*   **Neuron**: The fundamental building block; it receives inputs, processes them, and passes an output.
*   **Weightage (Weights)**: Parameters determining the importance or strength of the connection.
*   **Bias**: Offsets allowing the activation function to shift for better fitting.
*   **Input Layer**: The first layer receiving raw data.
*   **Hidden Layer**: Intermediate layers where mathematical transformations extract features.

### 2. The Perceptron
The simplest neural network, invented by Frank Rosenblatt in 1958.
*   **Mechanism**: Inputs are multiplied by weights, summed, and passed through an activation function to determine output (1 for 'fire', 0 for silence).
*   **Cool Historical Fact**: It wasn't just code! It was originally constructed as a physical machine called the **Mark 1 Perceptron**. While limited to linear problems, it is the ancestor of all modern Deep Learning.

### 3. The Mathematical Formula of a Neuron
This is the core calculation performed by every unit in the network:

1. **Linear Summation**: $z = \sum (w_i \cdot x_i) + b$
2. **Activation Application**: $Output = \sigma(z)$

**Combined Formula:**
$$Output = \sigma\left(\sum_{i=1}^{n} w_i x_i + b\right)$$

### 4. Multi-Layer Perceptron (MLP)
To solve complex, non-linear tasks (like the XOR problem), we stack neurons into layers.
*   **Evolution**: Moving from a single unit to a dense web of parameters.
*   **Fully Connected (Dense)**: Every neuron in one layer connects to every neuron in the next.
*   **Non-Linearity**: Without non-linear activation, stacking layers would just be one big linear equation.

### 5. Common Activation Functions
*   **ReLU**: $f(x) = \max(0, x)$ (Efficient; outputs input if positive, else zero).
*   **Leaky ReLU**: $f(x) = \max(0.01x, x)$ (Prevents 'dead neurons' by allowing a small negative gradient).
*   **Sigmoid**: $f(x) = \frac{1}{1 + e^{-x}}$ (Maps values 0 to 1; used for binary classification).
*   **Tanh**: $f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$ (Maps values -1 to 1; zero-centered).
*   **Softmax**: Used in the final layer for multi-class probability distributions.

### 6. The Training Process
1.  **Forward Propagation**: Data moves from input to output to generate a prediction.
2.  **Loss Function**: Measures the 'distance' between prediction and reality (error).
3.  **Backward Propagation (Backprop)**: Calculating gradients to see which weights caused the error.
4.  **Optimizer**: Algorithms like **Adam** or **SGD** that update weights to reduce loss.

### 7. Advanced Architectures
*   **CNN (Convolutional)**: Specifically for computer vision using filters to scan images.
*   **Transformers**: The foundation for LLMs using **Self-Attention** to understand context.

## Practical Implementation: The NumPy Perceptron

Now we will apply the formula from Section 3: $Output = \sigma(\sum wx + b)$.

In this example, we use the **Heaviside Step Function** as our activation function $\sigma$: if the sum is greater than 0, it returns 1; otherwise, it returns 0.

```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, lr=1, epochs=10):
        # Initialize weights and bias to zeros (or random small numbers)
        self.W = np.zeros(input_size)
        self.b = 0
        self.lr = lr
        self.epochs = epochs

    def activation_fn(self, x):
        # Simple Step Function
        return 1 if x >= 0 else 0

    def predict(self, x):
        # The core formula: z = sum(w*x) + b
        z = self.W.dot(x) + self.b
        return self.activation_fn(z)

    def fit(self, X, d):
        # Basic training loop using the perceptron learning rule
        for _ in range(self.epochs):
            for i in range(d.shape[0]):
                y = self.predict(X[i])
                error = d[i] - y
                # Update weights and bias based on error
                self.W = self.W + self.lr * error * X[i]
                self.b = self.b + self.lr * error

# --- Demo: Teaching the Perceptron the 'AND' Logic Gate ---
# Inputs: [0,0], [0,1], [1,0], [1,1]
X = np.array([[0,0], [0,1], [1,0], [1,1]])
# Desired Outputs (AND gate: only 1 if both inputs are 1)
d = np.array([0, 0, 0, 1])

perceptron = Perceptron(input_size=2)
perceptron.fit(X, d)

print("Testing the trained Perceptron:")
for x in X:
    print(f"Input: {x} | Prediction: {perceptron.predict(x)}")
```
